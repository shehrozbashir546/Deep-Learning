```python
#check number of pictures in directory
def counter():
    dir_path = os.getcwd()
    print("The directory is ", dir_path)
    count = 0
# Iterate directory
    for path in os.listdir(dir_path):
        # check if current path is a file
        if os.path.isfile(os.path.join(dir_path, path)):
            count += 1
    print('File count:', count)
```
